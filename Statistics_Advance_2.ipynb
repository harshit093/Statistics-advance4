# **Theory Questions**

1. What is hypothesis testing in statistics?
  - Statistical hypothesis testing is a method used to determine if data supports a specific claim or hypothesis by analyzing a sample and comparing it to a null hypothesis, which assumes no effect or difference.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - A null hypothesis is a statement in which there is no relation between the two variables, while an alternative hypothesis is a statement in which there is some statistical relationship between the two variables.

3. What is the significance level in hypothesis testing, and why is it important?
  - In hypothesis testing, the significance level (often denoted as α) is the probability of rejecting a true null hypothesis, and it's crucial because it sets the threshold for determining statistical significance, guiding researchers in making informed decisions about their hypotheses.

4. what does a p-value represent in hypothesis testing?
  - In hypothesis testing, a p-value represents the probability of observing results as extreme as, or more extreme than, the ones obtained from a sample, assuming the null hypothesis is true.

5. how do you interpret the p-value in hypothesis testing?
  - In hypothesis testing, the p-value represents the probability of observing results as extreme as, or more extreme than, the ones obtained, assuming the null hypothesis is true.

6. what are type 1 and type 2 errors in hypothesis testing?
  - In hypothesis testing, a Type I error (false positive) occurs when you reject a true null hypothesis, while a Type II error (false negative) occurs when you fail to reject a false null hypothesis.

7. what is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - In hypothesis testing, a one-tailed test checks for a difference in one specific direction (e.g., greater than or less than), while a two-tailed test checks for a difference in either direction (greater than or less than).

8. What is the Z-test, and when is it used in hypothesis testing?
  - A z-test is a statistical test used in hypothesis testing to determine if there's a significant difference between a sample mean and a population mean, or between two sample means, when the population standard deviation is known or the sample size is large (n ≥ 30).

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - To calculate a z-score, you subtract the population mean from a data point and divide the result by the population standard deviation. In hypothesis testing, a z-score represents how many standard deviations a data point or sample mean is from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
  - The t-distribution is used as an alternative to the normal distribution when the population standard deviation is unknown and the sample size is small (typically less than 30). It's a more conservative distribution with heavier tails, accounting for the increased uncertainty in estimating the population standard deviation from a small sample.

11. What is the difference between a Z-test and a T-test?
  - The main difference between a Z-test and a T-test lies in the sample size and knowledge of the population standard deviation. Z-tests are used for large samples (n > 30) or when the population standard deviation is known, while T-tests are used for small samples (n < 30) or when the population standard deviation is unknown.

12. What is the T-test, and how is it used in hypothesis testing?
  - A t-test is a statistical test used to determine if there's a significant difference between the means of two groups or a sample mean and a known value, often used in hypothesis testing when the population standard deviation is unknown.

13. what is the relationship between z-test and t-test in hypothesis testing?
  - Z-tests and t-tests are both used in hypothesis testing to compare means, but they differ primarily in their assumptions about sample size and population standard deviation.

14. What is a confidence interval, and how is it used to interpret statistical results?
  - A confidence interval (CI) is a range of values used to estimate an unknown population parameter (like a mean) based on sample data, with a certain level of confidence (e.g., 95%) that the true value lies within that range.

15. What is the margin of error, and how does it affect the confidence interval?
  - Each interval reflects the range within which one may have 95% confidence that the true percentage may be found, given a reported percentage of 50%. The margin of error is half the confidence interval.

16. How is Bayes' Theorem used in statistics, and what is its significance?
  - Bayes' Theorem, a fundamental concept in probability and statistics, allows you to update the probability of a hypothesis based on new evidence, moving from a prior probability to a posterior probability.

17. What is the Chi-square distribution, and when is it used?
  - The chi-square distribution is a continuous probability distribution used in hypothesis testing, particularly for determining if observed data deviates significantly from expected data or if two categorical variables are related, and is defined by its degrees of freedom.

18. What is the Chi-square goodness of fit test, and how is it applied?
  - The chi-square goodness-of-fit test is a statistical test used to determine if a sample data set matches a hypothesized distribution by comparing observed frequencies to expected frequencies.

19. What is the F-distribution, and when is it used in hypothesis testing?
  - The F-distribution is a continuous probability distribution used in hypothesis testing, particularly when comparing variances between two or more populations, most notably in the context of ANOVA (Analysis of Variance) and F-tests.

20. What is an ANOVA test, and what are its assumptions?
  - ANOVA (Analysis of Variance) is a statistical test used to compare the means of two or more groups, analyzing variance within and between groups to determine if there are statistically significant differences. Its key assumptions include independent observations, normally distributed data, and homogeneity of variance (equal variances across groups).

21. what are the different types of anova tests?
  - The main types of ANOVA tests are one-way ANOVA (for one independent variable) and two-way ANOVA (for two or more independent variables), with extensions like repeated measures ANOVA and factorial ANOVA.

22. What is the F-test, and how does it relate to hypothesis testing?
  - The F-test is a statistical test that compares variances between two or more groups, often used in hypothesis testing to determine if there's a significant difference in means or to assess the overall significance of a regression model.

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):


    standard_error = population_std / np.sqrt(sample_size)

    z_statistic = (sample_mean - population_mean) / standard_error

    p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

    critical_z = stats.norm.ppf(1 - alpha / 2)

    reject_null = np.abs(z_statistic) > critical_z

    x = np.linspace(population_mean - 4 * standard_error, population_mean + 4 * standard_error, 200)
    y = stats.norm.pdf(x, population_mean, standard_error)

    plt.plot(x, y, label='Sampling Distribution of the Mean')
    plt.axvline(population_mean, color='black', linestyle='--', label='Population Mean')
    plt.axvline(population_mean + critical_z * standard_error, color='red', linestyle='--', label=f'Critical Z (+/- {critical_z:.2f})')
    plt.axvline(population_mean - critical_z * standard_error, color='red', linestyle='--')
    plt.axvline(sample_mean, color='green', linestyle='-', linewidth=2, label=f'Sample Mean ({sample_mean:.2f})')

    plt.fill_between(x, y, where=(x > population_mean + critical_z * standard_error), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x, y, where=(x < population_mean - critical_z * standard_error), color='red', alpha=0.3)

    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.title('Two-Tailed Z-Test Decision Region')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical Z-values: +/- {critical_z:.4f}")
    if reject_null:
        print(f"At a significance level of {alpha}, we reject the null hypothesis.")
    else:
        print(f"At a significance level of {alpha}, we fail to reject the null hypothesis.")

    return z_statistic, p_value, reject_null

if __name__ == "__main__":
    sample_mean = 105
    population_mean = 100
    population_std = 15
    sample_size = 30
    alpha = 0.05

    z_stat, p_val, reject = two_tailed_z_test(sample_mean, population_mean, population_std, sample_size, alpha)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.


def plot_type_errors(null_mean, null_std, alt_mean, alt_std, alpha=0.05):

    x = np.linspace(null_mean - 4 * null_std, null_mean + 4 * null_std, 1000)

    null_dist = stats.norm.pdf(x, loc=null_mean, scale=null_std)
    alt_dist = stats.norm.pdf(x, loc=alt_mean, scale=alt_std)

    critical_value = stats.norm.ppf(1 - alpha, loc=null_mean, scale=null_std)

    type_1_area = stats.norm.cdf(critical_value, loc=null_mean, scale=null_std)

    type_2_area = stats.norm.cdf(critical_value, loc=alt_mean, scale=alt_std)

    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist, label="Null Distribution (H₀)", color="blue")
    plt.plot(x, alt_dist, label="Alternative Distribution (H₁)", color="red")

    plt.fill_between(x, null_dist, 0, where=(x >= critical_value), color="orange", alpha=0.5, label=f"Type 1 Error Area (alpha = {alpha})")

    plt.fill_between(x, alt_dist, 0, where=(x <= critical_value), color="yellow", alpha=0.5, label="Type 2 Error Area")

    plt.axvline(critical_value, color="black", linestyle="--", label=f"Critical Value ({critical_value:.2f})")
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Test Statistic")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)

    plt.show()

    print(f"Type 1 Error (False Positive) Probability: {type_1_area:.4f}")
    print(f"Type 2 Error (False Negative) Probability: {type_2_area:.4f}")

null_mean = 0
null_std = 1
alt_mean = 1
alt_std = 1

plot_type_errors(null_mean, null_std, alt_mean, alt_std, alpha=0.05)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results

def independent_t_test(sample1, sample2, alpha=0.05):

    t_statistic, p_value = stats.ttest_ind(sample1, sample2)

    print("\n--- Independent Samples T-test Results ---")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Significance Level (Alpha): {alpha:.2f}")

    if p_value < alpha:
        print("\nInterpretation:")
        print(f"The p-value ({p_value:.4f}) is less than the significance level ({alpha:.2f}).")
        print("This suggests that there is statistically significant evidence to reject the null hypothesis.")
        print("We can conclude that there is a significant difference in the means of the two groups.")
    else:
        print("\nInterpretation:")
        print(f"The p-value ({p_value:.4f}) is greater than or equal to the significance level ({alpha:.2f}).")
        print("This suggests that there is not enough statistically significant evidence to reject the null hypothesis.")
        print("We cannot conclude that there is a significant difference in the means of the two groups based on this data.")

    return t_statistic, p_value

if __name__ == "__main__":

    group_a = np.array([25, 30, 35, 40, 45])
    group_b = np.array([20, 28, 32, 38, 42])

    t_stat, p_val = independent_t_test(group_a, group_b)

    print("\n--- Another Example with more data ---")
    np.random.seed(0)
    group_x = np.random.normal(loc=50, scale=10, size=100)
    group_y = np.random.normal(loc=55, scale=10, size=100)
    independent_t_test(group_x, group_y, alpha=0.01)
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import math

# Function to calculate margin of error
def margin_of_error(confidence_level, std_dev, sample_size):
    # Z-scores for common confidence levels
    z_scores = {
        90: 1.645,
        95: 1.96,
        99: 2.576
    }

    # Check if the confidence level is in the dictionary
    if confidence_level not in z_scores:
        return "Invalid confidence level. Choose from 90, 95, or 99."

    # Get the Z-score for the given confidence level
    z_score = z_scores[confidence_level]

    # Calculate margin of error
    moe = z_score * (std_dev / math.sqrt(sample_size))
    return moe

# Example usage
confidence_level = float(input("Enter confidence level (90, 95, or 99): "))
std_dev = float(input("Enter sample standard deviation: "))
sample_size = int(input("Enter sample size: "))

moe = margin_of_error(confidence_level, std_dev, sample_size)
print(f"Margin of Error: {moe:.4f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.


def bayesian_inference(prior_red, prior_green, likelihood_red_given_red, likelihood_red_given_green):

    evidence = (likelihood_red_given_red * prior_red) + (likelihood_red_given_green * prior_green)


    posterior_red_given_red = (likelihood_red_given_red * prior_red) / evidence
    posterior_red_given_green = (likelihood_red_given_green * prior_green) / evidence

    return posterior_red_given_red, posterior_red_given_green


prior_red = 0.5
prior_green = 0.5

likelihood_red_given_red = 0.7
likelihood_red_given_green = 0.3

posterior_red_given_red, posterior_red_given_green = bayesian_inference(
    prior_red, prior_green, likelihood_red_given_red, likelihood_red_given_green
)

print(f"Posterior Probability of Red Bag given Red Marble: {posterior_red_given_red:.4f}")
print(f"Posterior Probability of Green Bag given Red Marble: {posterior_red_given_green:.4f}")
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
import scipy.stats as stats

data = np.array([[30, 10],
                 [15, 25]])

chi2, p_value, dof, expected = stats.chi2_contingency(data)

print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies Table:\n{expected}")

alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis: The variables are not independent.")
else:
    print("Fail to reject the null hypothesis: The variables are independent.")
```

```python
#16. Implement an F-test using Python to compare the variances of two random samples.


def f_test(sample1, sample2, alpha=0.05):

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        F_statistic = var1 / var2
        df1 = len(sample1) - 1
        df2 = len(sample2) - 1
    else:
        F_statistic = var2 / var1
        df1 = len(sample2) - 1
        df2 = len(sample1) - 1

    p_value = 2 * min(stats.f.cdf(F_statistic, df1, df2), 1 - stats.f.cdf(F_statistic, df1, df2))

    print(f"F-statistic: {F_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")

    return F_statistic, p_value

np.random.seed(42)

sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

f_test(sample1, sample2)
```

```python
#21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.


def plot_f_distribution(dfn, dfd):

    x = np.linspace(0, 5, 1000)

    pdf = stats.f.pdf(x, dfn, dfd)

    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf, label=f'F-distribution (dfn={dfn}, dfd={dfd})', color='blue')

    critical_value = stats.f.ppf(0.95, dfn, dfd)
    plt.axvline(critical_value, color='red', linestyle='--', label=f'Critical value (95% CI): {critical_value:.2f}')

    plt.fill_between(x, 0, pdf, where=(x > critical_value), color='red', alpha=0.3)

    plt.title(f"F-distribution (dfn={dfn}, dfd={dfd})")
    plt.xlabel("F-statistic value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_f_distribution(dfn=3, dfd=15)
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.


n = 25
s_squared = 50
sigma_squared_0 = 45
alpha = 0.05

df = n - 1

chi_square_statistic = (df * s_squared) / sigma_squared_0

chi_square_critical_low = stats.chi2.ppf(alpha / 2, df)
chi_square_critical_high = stats.chi2.ppf(1 - alpha / 2, df)

print(f"Chi-square test statistic: {chi_square_statistic:.4f}")
print(f"Critical values: Lower = {chi_square_critical_low:.4f}, Upper = {chi_square_critical_high:.4f}")

if chi_square_statistic < chi_square_critical_low or chi_square_statistic > chi_square_critical_high:
    result = "Reject the null hypothesis"
else:
    result = "Fail to reject the null hypothesis"

print(result)
```

```python
#25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

def z_test_proportions(x1, n1, x2, n2, alpha=0.05):

    p1 = x1 / n1
    p2 = x2 / n2

    pooled_p = (x1 + x2) / (n1 + n2)

    SE = math.sqrt(pooled_p * (1 - pooled_p) * (1/n1 + 1/n2))

    Z = (p1 - p2) / SE

    p_value = 2 * (1 - stats.norm.cdf(abs(Z)))

    print(f"Sample 1 Proportion (p1): {p1:.4f}")
    print(f"Sample 2 Proportion (p2): {p2:.4f}")
    print(f"Pooled Proportion (P): {pooled_p:.4f}")
    print(f"Standard Error (SE): {SE:.4f}")
    print(f"Z statistic: {Z:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at alpha={alpha}. The proportions are significantly different.")
    else:
        print(f"Fail to reject the null hypothesis at alpha={alpha}. No significant difference between proportions.")

x1 = 45
n1 = 100
x2 = 55
n2 = 120

z_test_proportions(x1, n1, x2, n2, alpha=0.05)
```

```python
#27.  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

np.random.seed(42)
observed_data = np.random.choice(['A', 'B', 'C', 'D'], size=200, p=[0.2, 0.3, 0.4, 0.1])

observed_counts = [np.sum(observed_data == category) for category in ['A', 'B', 'C', 'D']]

expected_probabilities = [0.25, 0.25, 0.25, 0.25]
expected_counts = [len(observed_data) * p for p in expected_probabilities]

chi2_stat, p_value = stats.chisquare(observed_counts, expected_counts)

print(f"Observed Counts: {observed_counts}")
print(f"Expected Counts: {expected_counts}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print(f"Reject the null hypothesis at alpha={alpha}. The observed data does not follow the expected distribution.")
else:
    print(f"Fail to reject the null hypothesis at alpha={alpha}. The observed data follows the expected distribution.")

categories = ['A', 'B', 'C', 'D']
x = np.arange(len(categories))

fig, ax = plt.subplots()
bar_width = 0.35

ax.bar(x - bar_width/2, observed_counts, bar_width, label='Observed', color='blue')
ax.bar(x + bar_width/2, expected_counts, bar_width, label='Expected', color='red')

ax.set_xlabel('Categories')
ax.set_ylabel('Counts')
ax.set_title('Observed vs Expected Counts for Chi-square Test')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```
